```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from funcoes_gerais import *
```

```python
lista_obj = []
lista_tempos = []
with open("objetivos_GA_parametrizado1.pkl", "rb") as infile:
    dados1 = pickle.load(infile)
    objetivos_pandas = pd.DataFrame(dados1)
    objetivos_pandas.iloc[20]
    objetivos_unstack = objetivos_pandas.iloc[20].unstack(level=-1)
    objetivos_unstack
    lista_obj.append(objetivos_unstack.transpose())
    

with open("tempos_pandas_parametrizado1.pkl", "rb") as infile:
    tempos1 = pickle.load(infile) 
    tempos_pandas = pd.Series(tempos1)
    tempos_unstack = tempos_pandas.unstack(level=-1)
    lista_tempos.append(tempos_unstack.transpose())      
```

```python
conjuntos = [10, 20, 50, 100, 200, 500, 1000]
lista_hs = [0.2,0.4,0.6,0.8]
lista_crossover = [0 , 1]
lista_elitismo = [0.35 , 0.75 , 1]
lista_so_filho_mutante = [0 , 1]
lista_taxa_mutacao = [0.2 , 1 , 5]
objetivos_min = {}
repeticoes = 4
for conjunto in conjuntos:
    for h in lista_hs:
        for problema in range(1, 11):
            obj_min = 99999999999
            for i in range(0, repeticoes):
                objetivos_min[(conjunto, problema,h, i)] = min(dados1[(conjunto, h, problema, i)]) 
```

```python
tempos = {}
for conjunto in conjuntos:
    for h in lista_hs:
        for problema in range(1, 11):
            tempos[(conjunto, problema,h)] = np.mean([tempos1[(conjunto, h,problema,i)] for i in range(0,repeticoes)])
```

```python
objetivos_media = {}
for conjunto in conjuntos:
    for h in lista_hs:
        for problema in range(1, 11):
            objetivos_media[(conjunto, problema,h)] = np.mean([objetivos_min[(conjunto, problema,h,i)] for i in range(0,repeticoes)])
```

```python
biskup = pd.read_excel('../objetivos_biskup.xlsx').set_index(["Unnamed: 0","Unnamed: 1", "Unnamed: 2"])
objetivos_biskup = biskup.to_dict()["objetivo"]
resultados = {}

for key in objetivos_biskup:
    resultados[key] = (objetivos_media[key] - objetivos_biskup[key])/objetivos_biskup[key]
```
